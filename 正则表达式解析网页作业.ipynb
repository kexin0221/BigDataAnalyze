```python
# 用search方法匹配出“'abc45'”字符串中的“45”
import re

s = "'abc45'"
match = re.search(r'\d+', s)
match.group()
```

```python
# 用findall方法查找出'ab2c3ed'字符串中的数字
s = 'ab2c3ed'
result = re.findall(r'\d', s)
result
```

```python
# 网页实例
import requests
import chardet
url = 'https://www.tipdm.com'
ua = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
rqg = requests.get(url, headers=ua)
rqg.encoding = chardet.detect(rqg.content)['encoding']
```

```python
# 使用search()方法查找title中的内容,并输出
title_pattern = r'<title>(.*?)</title>'
match = re.search(title_pattern, rqg.text)
match.group(1)
```

```python
# 使用findall()方法查找title中的内容,并输出
titles = re.findall(r'<title>(.*?)</title>', rqg.text)
for title in titles:
    print(title)
```
