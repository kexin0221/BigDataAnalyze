```python
# 代码3-12
import requests
# 导入etree模块
from lxml import etree
url = 'https://www.tipdm.com'
ua = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
rqg = requests.get(url, headers=ua)
rqg.encoding = 'utf-8'
# 初始化HTML对象
html = etree.HTML(rqg.text, parser=etree.HTMLParser(encoding='utf-8'))
# 输出修正后的HTML对象（如有必要）
result = etree.tostring(html, encoding='utf-8', pretty_print=True, method='html')
print('修正后的HTML对象：', result)
# print('格式化后的HTML对象：', result.decode('utf-8')) = target[0].xpath('string(.)').strip()  # strip方法可用于去除多余的空格
target_text = result.decode('utf-8')
print('ul节点下的全部文本内容：', target_text)
```

```python
# 代码3-14
# 通过名称定位head节点
result0 =  html.xpath('head')
print('名称定位head节点结果：', result0)

# 按节点层级定位title节点
result1 = html.xpath('/html/head/title')
print('节点层级定位结果：', result1)

# 通过名称定位title节点
result2 = html.xpath('head/title')
print('名称定位title节点结果：', result2)

# 通过搜索定位title节点
result3 = html.xpath('//title')
print('搜索定位title节点结果：', result3)
```

```python
# 代码3-15
# 定位header节点
result1 = html.xpath('//header[@class="header"]')
print('class属性定位结果：', result1)

# 定位ul节点，并输出id属性定位结果
result2 = html.xpath('//ul[@id="nav"]')
print('id属性定位结果：', result2)
```

```python
# 代码3-16
# 定位并获取title节点的文本内容
title = html.xpath('//title/text()')
print('title节点的文本内容：', title)
```

```python
# 代码3-17
# 定位属性id值以me开头的ul节点，并提取其所有子孙节点a内的文本内容
content = html.xpath('//ul[starts-with(@id, "me")]//a/text()')
for i in content:
    print(i)

# 提取对应链接
url_list = html.xpath('//ul[starts-with(@id, "me")]//a/@href')
for i in url_list:
    print(i)

# 定位属性id值以me开头的ul节点
target = html.xpath('//ul[starts-with(@id, "me")]')

# 提取该节点下的全部文本内容
target_text = target[0].xpath('string(.)').strip()       # strip方法可用于去除多余的空格
print('ul节点下的全部文本内容：', target_text)
```
