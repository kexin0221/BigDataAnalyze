#### 2.Beautiful_findall

```python
for p in range(1, 6):
    print(f"第{p}页")
    resp = requests.get(url_template.format(p), headers=headers)
    resp.encoding = 'utf-8'
    soup = BeautifulSoup(resp.text, 'lxml')
    links = soup.find_all('a', target='_blank')

    for a in links:
        title = a.get_text(strip=True)
        href = a.get('href', '')

        if "zqrb.cn" in href and len(title) > 10:
            print(f"{title}\n{href}")
```

#### 3.beautiful_select

```python
for p in range(1, 6):
    print(f"第{p}页")
    resp = requests.get(url_template.format(p), headers=headers)
    resp.encoding = 'utf-8'
    soup = BeautifulSoup(resp.text, 'lxml')

    for a in soup.select('a[target="_blank"]'):
        title = a.get_text(strip=True)
        href = a.get('href', '')

        if "zqrb.cn" in href and len(title) > 10:
            print(f"{title}\n{href}")
```
